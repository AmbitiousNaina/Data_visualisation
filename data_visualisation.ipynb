```python
Year=[2012,2014,2017,2019,2022]
India=[58.5, 70, 89.12, 102.13, 200]
Bangladesh=[49.3,65.5, 79.45, 102,167.70]
```

```python
plt.ylabel("India")
plt.xlabel("Years")
plt.plot(Year,India, color = 'red')
```

```python
plt.ylabel("Bangladesh")
plt.xlabel("Years")
plt.plot(Year,Bangladesh, color = 'green')
```

```python
plt.plot(Year,India, color = 'orange')
plt.plot(Year,Bangladesh, color = 'green')
```

```python
plt.plot(Year,India, color = 'orange',label = 'India')
plt.plot(Year,Bangladesh, color = 'green', label = 'Bangladesh')
plt.legend()
plt.show()
```

```python
plt.plot(Year,India, color = 'orange',label = 'India', marker = 'o')
plt.plot(Year,Bangladesh, color = 'green', label = 'Bangladesh', marker = 's')
plt.legend()
plt.show()
```

```python
from matplotlib.markers import MarkerStyle
```

```python
plt.plot(Year,India, color = 'orange',label = 'India', marker = 'o', markersize = 12)
plt.plot(Year,Bangladesh, color = 'green', label = 'Bangladesh', marker = 's', markersize = 10)
plt.legend()
plt.show()
```

```python
plt.plot(Year,India, color = 'orange',label = 'India', marker = 'o', markersize = 12, mec = 'b')
plt.plot(Year,Bangladesh, color = 'green', label = 'Bangladesh', marker = 's', markersize = 10)
plt.legend()
plt.show()
```

```python
plt.plot(Year,India, color = 'orange',label = 'India', marker = 'o', markersize = 12, mec = 'b')
plt.plot(Year,Bangladesh, color = 'green', label = 'Bangladesh', marker = 's', markersize = 10, mec='black', mfc= 'b')
plt.legend()
plt.show()
```

```python
plt.scatter(x,y,label='Marks', c = 'g')
plt.title("Scatter plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.savefig('Marks.png')
plt.show()
```

```python
y = x * x
plt.plot(x, y, 'r*', linestyle='dashed', linewidth=2, markersize=10)
plt.title("2-D line plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(y,x,'g--')
plt.subplot(2,2,3)
plt.plot(x,y,'b--')
plt.subplot(2,2,4)
plt.plot(x,y, 'b*')
```

```python
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0, 0.2, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.2f%%', shadow=True)

plt.axis('equal')
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
data = [1, 2, 3, 4, 5, 6, 7, 8]
df = pd.DataFrame({'Value': data})

# Create violin plot
sns.violinplot(data=df)


# Adding title and labels
plt.title("Violin Plot ")
plt.xlabel("Day")
plt.ylabel("Total")

# Displaying the plot
plt.show()
```
